```python
import pandas as pd
```

```python
df = pd.read_csv('cleaneddataset.csv')
df_description = pd.read_csv('symptom_Description.csv')
df_precaution = pd.read_csv('symptom_precaution.csv')
```

```python
df.head()
```

```python
df_description.head()
```

```python
df_precaution.head()
```

```python
#uniformity for consistency
df_description = df_description.apply(lambda x: x.str.lower() if x.dtype == "object" else x)
df_precaution = df_precaution.apply(lambda x: x.str.lower() if x.dtype == "object" else x)
```

```python
# Merge symptom description into the dataset
df_mergedd = pd.merge(df, df_description, on="Disease", how="left")
df_mergedd.head()
```

```python
# Merge symptom precaution into the the dataset
df_mergedp = pd.merge(df_mergedd, df_precaution, on="Disease", how="left")
df_mergedp.head()
```

```python
# Check missing values
print(df_mergedp.isnull().sum())
```

```python
#Check for duplicates
print(df_mergedp.duplicated().sum())  
```

```python
df_mergedp.head()
```

```python
df_mergedp.to_csv('mergeddata.csv', index=False)
```
